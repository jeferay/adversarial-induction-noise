```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.figure(figsize=(4,3))
sns.set_style("whitegrid")
label_size=9
legend_size=9


df=pd.DataFrame(data={
    "threshold":['0.9','0.99','0.999'],
    "rob":[8.44,0.57,0.57],
    "nat":[64.21,56.52,56.52]
    
})
plt.xlim([0,2])
#plt.ylim([0,0.6])
sns.lineplot(data=pd.melt(df, ['threshold']), x='threshold', y='value', hue='variable',lw=2)
ax = plt.gca()
ax.legend(['rob','nat'], loc='center right',fontsize=legend_size)

#ax.get_legend().remove()
plt.xticks(size=9)
ax.set_xlabel('PSR Threshold',fontsize=label_size)
ax.set_ylabel('Test Accuracy',fontsize=label_size)

#plt.title('Adversarial Training',fontsize=15)

plt.savefig('figures/insensitive_to_threshold.pdf', bbox_inches='tight',  pad_inches = 0)
plt.show()

```
