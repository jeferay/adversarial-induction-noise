```python
import seaborn as sns
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
plus_one = torch.load('confusion_matrix/plus_one.pth')
cifar10 = torch.load('confusion_matrix/CIFAR10.pth')
near_swap = torch.load('confusion_matrix/near_swap.pth')
plus_one = np.array(plus_one/plus_one.sum(1) * 100)
cifar10 = np.array(cifar10/cifar10.sum(1) * 100)
near_swap = np.array(near_swap/near_swap.sum(1)*100)

```

```python

text_size = 9

plt.figure(figsize=(17,5))
plt.subplot(1,3,1)
ax = plt.gca()
sns.heatmap(data=plus_one,square=True,annot=True,cbar=True,cmap="Oranges",annot_kws={'size': text_size })
plt.title(label='ADVIN-NextCycle')
plt.xlabel('Predicted class')
plt.ylabel('Ground truth label')
ax.set_xticklabels([str(i) for i in range(1,11)])
ax.set_yticklabels([str(i) for i in range(1,11)])


plt.subplot(1,3,2)
ax = plt.gca()
sns.heatmap(data=cifar10,square=True,annot=True,cbar=True,cmap="Oranges",annot_kws={'size': text_size})
plt.title(label='Clean')
plt.xlabel('Predicted class')
plt.ylabel('Ground truth label')
ax.set_xticklabels([str(i) for i in range(1,11)])
ax.set_yticklabels([str(i) for i in range(1,11)])

plt.subplot(1,3,3)
ax = plt.gca()
sns.heatmap(data=near_swap,square=True,annot=True,cbar=True,cmap="Oranges",annot_kws={'size': text_size})
plt.title(label='ADVIN-NearSwap')
plt.xlabel('Predicted class')
plt.ylabel('Ground truth label')
#0lt.xticks(range(1,11))
ax.set_xticklabels([str(i) for i in range(1,11)])
ax.set_yticklabels([str(i) for i in range(1,11)])

plt.savefig('figures/heatmap.pdf', bbox_inches='tight',  pad_inches = 0)
plt.show()
```
