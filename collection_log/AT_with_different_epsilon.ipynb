```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.figure(figsize=(8,3))
sns.set_style("whitegrid")
label_size=9
legend_size=7

plt.subplot(1,2,1)

df=pd.DataFrame(data={
    "epsilon":['0','2','4','8','16','24','32','64'],
    "at":[0,0.2926,0.4105,0.5171,0.5261,0.4046,0.2934,0.10]
    
})
plt.xlim([0,7])
plt.ylim([0,0.6])
sns.lineplot(data=pd.melt(df, ['epsilon']), x='epsilon', y='value', hue='variable',lw=2)
ax = plt.gca()
ax.legend(['AT'], loc='center right',fontsize=legend_size)
ax.get_legend().remove()
plt.xticks(size=9)
ax.set_xlabel('Epsilon',fontsize=label_size)
ax.set_ylabel('Test Robust Accuracy',fontsize=label_size)



#plt.title('Adversarial Training',fontsize=15)



plt.savefig('figures/AT_with_different_epsilon.pdf', bbox_inches='tight',  pad_inches = 0)
plt.show()

```
